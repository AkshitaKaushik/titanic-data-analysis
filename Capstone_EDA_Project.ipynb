# **Project Summary -**

This project focuses on **Exploratory Data Analysis (EDA)** and **Data Visualization** using the Titanic dataset. The primary objective of the project was to understand the dataset, clean and preprocess the data, identify patterns and relationships among different variables, and draw meaningful insights through various visualizations. The Titanic dataset is one of the most widely used datasets in data science and machine learning because it contains information about passengers such as age, gender, passenger class, fare, family members, embarkation port, cabin details, and whether they survived the disaster.

The first stage of the project involved understanding the dataset and its features. Each column was analyzed to determine its purpose and significance. Special attention was given to features such as Pclass, Sex, Age, Fare, Embarked, SibSp, Parch, and Cabin. During the data wrangling process, missing values and inconsistent data were identified and handled appropriately. Missing values in the Age column were replaced using the median value, while missing values in the Embarked column were filled using the most frequent embarkation port (mode). Since the Cabin column contained a large number of missing values, the deck information was extracted by taking the first character of the cabin number, and missing entries were labeled as "Unknown" to preserve potentially useful information. The original Cabin column was then removed. Additionally, a new feature called FamilySize was created by combining the SibSp and Parch columns along with the passenger themselves, providing a better representation of family size aboard the ship. The Sex column was encoded into numerical values to prepare the dataset for future machine learning applications.

After cleaning the data, exploratory data analysis was performed using various visualization techniques. Different chart types such as bar charts, pie charts, histograms, scatter plots, box plots, line plots, stacked bar charts, and heatmaps were used to explore relationships between variables. Visualizations were created to analyze the effect of gender, passenger class, fare, age, family size, embarkation port, and deck on survival. Histograms helped understand the distribution of passenger age and fare, while scatter plots illustrated relationships between numerical variables. Box plots were useful for identifying differences in fare and age across survival groups and passenger classes. Pie charts displayed the proportion of passengers belonging to each passenger class, and heatmaps highlighted correlations among numerical features.

Several valuable insights were obtained during the analysis. Female passengers had a significantly higher survival rate than male passengers. First-class passengers were more likely to survive compared to those in lower classes, indicating that socioeconomic status influenced survival. Passengers who paid higher fares generally experienced better survival outcomes. Family size also showed an impact, with passengers traveling in small families tending to have higher survival rates than those traveling alone or in very large groups. The analysis of embarkation ports and deck information provided additional context regarding passenger demographics and survival patterns. Correlation analysis further revealed meaningful relationships among numerical variables such as fare, passenger class, family size, and survival.

Overall, this project successfully demonstrated the importance of data preprocessing, feature engineering, exploratory data analysis, and visualization in extracting meaningful information from raw data. The findings provide a clear understanding of the factors that influenced passenger survival and highlight how effective data analysis can support better decision-making. This project also strengthened practical skills in Python, Pandas, Matplotlib, and Seaborn while building a strong foundation for future machine learning and predictive modeling tasks.

## 3. ***Data Wrangling***

### Data Wrangling Code

The 'Age' column contains 177 missing values. So, replacing it with the median value.

````python
df['Age'].median()
````

````python
df['Age'].fillna(df['Age'].median(), inplace=True)
````

'Cabin' has 77% of the null values. So, extracting the 'Deck' from the Cabin number.

````python
# Creating a new column named 'Deck'
df['Deck']=df['Cabin'].str[0]
df.head()
````

````python
# Replacing the NaN values with Unknown, so that, it can be used in future for graphical representation (and we do not ignore the missing values).
df['Deck'].fillna('Unknown', inplace=True)
````

````python
df.head()
````

````python
# Now, dropping the column 'Cabin' because we have extracted all needed information from it.
df.drop(columns='Cabin', inplace=True)
# Or, can use: df.drop("Cabin", axis=1, inplace=True)
````

````python
# The 'Cabin' column has been removed/dropped.
df.head()
````

Filling missing 'Embarked' values with the most frequent port.

````python
df['Embarked'].mode()
````

````python
df['Embarked']=df['Embarked'].fillna(df['Embarked'].mode()[0])
print(df['Embarked'].isnull().sum())
````

Creating a new column name 'FamilySize' and eliminating two different columns.

````python
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
df.drop(columns=['SibSp','Parch'], inplace=True)
df.head()
````

Dropping column 'PassengerId' because it isn't useful for future purposes.

````python
df.drop("PassengerId", axis=1, inplace=True)
df.head()
````

### What all manipulations have you done and insights you found?

Data wrangling refers to cleaning of data, for making further data analysis easier.
1. The missing values in the Age column, has been replaced by its median().
2. Cabin had more than 70% of its values as null values. So, the deck has been extracted from the cabin number. All the NaN has been replaced by 'Unknown' category. And, the Cabin column has been dropped.
3. Embarked consisted of 2-3 null values. They have been replaced by the most common port, by the use of mode().
4. A new column 'FamilySize' has been created, which represents the total number of family members, which boarded the ship. This, in turn, helped me to reduce no. of columns, and make it easier to analyse the given data.
5. The column 'PassengerId' has been dropped because it will be no longer used in future. Thus, it also has no meaningful relationship with survival.
